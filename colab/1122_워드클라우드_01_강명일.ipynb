## 1.  AI 워드 클라우드

```python
# 라이브러리 가져오기

# 위키피디아 사용을 위한 라이브러리
import wikipedia
# 그래프 및 시각화를 위한 라이브러리
import matplotlib.pyplot as plt
# 워드클라우드 생성을 위한 라이브러리
from wordcloud import WordCloud, STOPWORDS
```

```python
# 위키피디아에서 컨텐츠 제목을 가져온다.
wiki = wikipedia.page('Artificial intelligence')
# 이 페이지의 텍스트 컨텐츠를 추출한다.
text = wiki.content
```

```python
# 텍스트를 가지고 워드클라우드 생성.
wc = WordCloud(width = 400, height = 400).generate(text)

# 새로운 figure를 생성한다.
plt.figure(figsize=(10,10))

# wc를 이미지로 표시
plt.imshow(wc)

# 그래프를 화면에 출력
plt.show()

# 워드클라우드 저장
wc.to_file('AI_클라우드1.png')
```

```python
# 중지어가 제외된 워드 클라우드 만들기
# 집합 => 중복을 허용하지 않는다. => union(교집합)
s_words = STOPWORDS.union({'one', 'two', 'using', 'output', 'way', 'tool', 'step', 'use', 'make', 'area', 'well', 'layer', 'AI'})

# 텍스트를 가지고 워드클라우드 생성.
wc = WordCloud(width = 400, height = 400, stopwords = s_words).generate(text)

# 새로운 figure를 생성한다.
plt.figure(figsize=(10,10))

# wc를 이미지로 표시
plt.imshow(wc)

# 그래프를 화면에 출력
plt.show()

# 워드클라우드 저장
wc.to_file('AI_클라우드2.png')
```
